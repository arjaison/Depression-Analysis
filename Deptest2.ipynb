```
import pandas as pd
import pickle as pickle  

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from lightgbm import LGBMClassifier    
from sklearn.metrics import accuracy_score, f1_score
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df2 = pd.read_csv(io.BytesIO(uploaded['datasetchange.csv']))
```

```
df2.head(5)
```

```
print('Columns of dataset \n',df2.columns)
```

```
print('column Unique_column_count')
for col in df2:
    print(len(df2[col].unique()), col)
```

```
dropping_cols = ['What is your name?','Which school are you studying in?','What is your favourite food?','How do you workout everyday?','How do you start your day?','How do you usually feel after a nap?', 'Mention a health problem which troubles you often']
df2.drop(dropping_cols, axis=1, inplace=True)
```

```

print('column Unique_column_values')
for col in df2:
    print( df2[col].unique(), col)
```

```
dataset_features = df2.drop(['Result'], axis=1)
dataset_target = pd.DataFrame(df2['Result'])
```

```
dataset_features.head()
```

```
dataset_features.dtypes
```

```
le = LabelEncoder()
```

```
for col in dataset_features:
    dataset_features[col] = le.fit_transform(dataset_features[col])
```

```
dataset_features.dtypes
```

```
dataset_features
```

```
dataset_features.to_csv('inout.csv')
```

```
files.download('inout.csv')
```

```
for col in dataset_features.columns:
  dataset_features[col]=dataset_features[col].apply(int)
```

```
dataset_features.dtypes
```

```

x_train, x_test, y_train, y_test = train_test_split(dataset_features, dataset_target, test_size=0.25, random_state=42)
```

```

print ('Train-set shape',x_train.shape, y_train.shape)
print ('Test-set shape', x_test.shape, y_test.shape)
```

```
model = LGBMClassifier(n_estimators=20)
```

```
model_fit = model.fit(x_train, y_train.values.ravel())
```

```
model_file = 'modelfinchange.pkl'
```

```
with open(model_file, 'wb') as fp:
     pickle.dump(model_fit, fp)
```

```
model_file = 'modelfinchange.pkl'
```

```
with open(model_file, 'rb') as fp:
     model_fit = pickle.load(fp)
```

```
y_predict = model_fit.predict(x_test)
print (y_predict)
```

```
print ('Accuracy:', accuracy_score(y_test, y_predict))
print ('F1-score:', f1_score(y_test, y_predict, average='macro'))
```
